# Automated real-time monitoring of taxi data

```python
import requests
import schedule
import time
import pandas as pd
from datetime import datetime

import folium 
import webbrowser
from folium.plugins import HeatMap

latitude = 1.3521
longitude = 103.8198

#latitude = 1.2966
#longtitude = 103.7764

#traffic_map = folium.Map(location=[latitude, longitude], zoom_start=11.3)
#traffic_map



headers = {
        "accept": "application/vnd.geo+json"
        }
url ="https://api.data.gov.sg/v1/transport/taxi-availability"

timestamp = 1545730073


traffic_map = folium.Map(location=[latitude, longitude], zoom_start=11.3)




def api():
    
    #traffic_map
    #etxract the data
    resp = requests.get(url , headers=headers ) # pass some header along with request
    data = resp.json() #to conver resp into dict      
    feat= data["features"][0]
    df = pd.DataFrame( feat['geometry']['coordinates'] )
    df.columns = ['x_coordinate','y_coordinate']
#     dt_object = datetime.fromtimestamp(timestamp)
    df.to_csv("latest"+"_"+str(datetime.now())+"_.csv")

    heat_data = [[row[-1],row[0]] for index, row in df.iterrows()]
    HeatMap(heat_data).add_to(traffic_map)
    
    display(traffic_map)


schedule.every(10).seconds.do(api)


while True:
    schedule.run_pending()
    time.sleep(1)

#traffic_map.save('mymap+str(datetime.now()).html')



```
